# Remove Duplicates within a span of a month

```python
# Deduplicate rows per (Buyer Company Name, Part Number, Condition Code) within ~30 days
# Keep at least one row (the earliest chronologically) per 30-day cluster.
import pandas as pd
from IPython.display import display

df = Rotabull_df.copy()
key_cols = ['Buyer Company Name', 'Part Number', 'Condition Code']
for c in key_cols + ['Received At (UTC)']:
    if c not in df.columns:
        raise KeyError(f"Missing expected column: {c}")

# Normalize and parse Received At into a datetime column
df['_recv_dt'] = pd.to_datetime(df['Received At (UTC)'], errors='coerce', utc=True)
# convert to naive UTC timestamps for easier arithmetic
if df['_recv_dt'].dt.tz is not None:
    df['_recv_dt'] = df['_recv_dt'].dt.tz_convert('UTC').dt.tz_localize(None)

# Prepare container for kept indices
kept_indices = []

# Iterate groups defined by keys
grouped = df.groupby(key_cols, sort=False)
for name, group in grouped:
    # Sort by Received datetime ascending
    g = group.sort_values('_recv_dt')
    last_kept_dt = None
    last_kept_idx = None
    for idx, row in g.iterrows():
        dt = row['_recv_dt']
        if pd.isna(dt):
            # If no date, keep only the first encountered (to ensure at least one copy kept)
            if last_kept_dt is None:
                kept_indices.append(idx)
                last_kept_dt = pd.NaT
                last_kept_idx = idx
            else:
                # skip additional undated rows
                continue
        else:
            if last_kept_dt is None or pd.isna(last_kept_dt):
                # Keep the first dated row
                kept_indices.append(idx)
                last_kept_dt = dt
                last_kept_idx = idx
            else:
                # If this row is within 30 days of last_kept_dt, consider it duplicate and skip
                if (dt - last_kept_dt) <= pd.Timedelta(days=30):
                    # duplicate within 30 days -> skip
                    continue
                else:
                    # far enough, keep and update last_kept_dt
                    kept_indices.append(idx)
                    last_kept_dt = dt
                    last_kept_idx = idx

# Build deduped and duplicates DataFrames
keep_mask = df.index.isin(kept_indices)
Checked_DF_deduped = df.loc[keep_mask].copy().reset_index(drop=True)
duplicates_found = df.loc[~keep_mask].copy().reset_index(drop=True)

# Drop helper column before exposing results
for d in (Checked_DF_deduped, duplicates_found):
    if '_recv_dt' in d.columns:
        d.drop(columns=['_recv_dt'], inplace=True, errors='ignore')

# Summary
total = len(df)
kept = len(Checked_DF_deduped)
removed = len(duplicates_found)
print(f"Total rows: {total:,}")
print(f"Kept rows after 30-day deduplication: {kept:,}")
print(f"Rows marked as duplicates (removed): {removed:,}")

if removed:
    print('\nSample duplicates (first 10):')
    display(duplicates_found.head(10))

print('\nSample kept rows (first 10):')
display(Checked_DF_deduped.head(10))

# Expose variables for later cells
# - Checked_DF_deduped: deduplicated DataFrame
# - duplicates_found: rows that were considered duplicates within 30 days

print('\nDeduplication complete. Use `Checked_DF_deduped` for exports or further analysis.')

Rotabull_df = Checked_DF_deduped.copy()
```

# Finding Inventory based on Date 

```python
def to_date_only(series):
    # parse to datetime (coerce invalids to NaT) then keep only the date part
    dt = pd.to_datetime(series, errors="coerce", infer_datetime_format=True)
    return dt.dt.date

# Rotabull copy: Received At (UTC) -> Received_Date (date only)
if "Received At (UTC)" in Rotabull_df_copy_price.columns:
    Rotabull_df_copy_price["Received_Date"] = to_date_only(Rotabull_df_copy_price["Received At (UTC)"])
    print("Rotabull: Created 'Received_Date' (date only). Nulls:", Rotabull_df_copy_price["Received_Date"].isna().sum())
else:
    print("Rotabull: column 'Received At (UTC)' not found")

# DF_Inventory copy: SNAP_DATE -> SNAP_DATE_date (date only)
if "SNAP_DATE" in DF_Inventory_copy.columns:
    DF_Inventory_copy["SNAP_DATE_date"] = to_date_only(DF_Inventory_copy["SNAP_DATE"])
    print("DF_Inventory: Created 'SNAP_DATE_date' (date only). Nulls:", DF_Inventory_copy["SNAP_DATE_date"].isna().sum())
else:
    print("DF_Inventory: column 'SNAP_DATE' not found")

# Quick sanity checks (show a few values)
print("Rotabull sample dates:", Rotabull_df_copy_price["Received_Date"].dropna().unique()[:5])
print("DF_Inventory sample dates:", DF_Inventory_copy["SNAP_DATE_date"].dropna().unique()[:5])
```

```python
Rotabull_df_copy_price["Part_norm"] = Rotabull_df_copy_price["Part Number"].astype(str).str.strip()
DF_Inventory_copy["PN_norm"] = DF_Inventory_copy["P/N"].astype(str).str.strip()

# ensure date-only columns exist and are actual date objects
# (assumes the to_date_only step ran earlier and created Received_Date and SNAP_DATE_date)
# coerce Qty Available to numeric for aggregation
DF_Inventory_copy["Qty Available"] = pd.to_numeric(DF_Inventory_copy.get("Qty Available"), errors="coerce")

# aggregate inventory by part and date (sum quantity available for that day)
inv_daily = (
    DF_Inventory_copy
    .dropna(subset=["PN_norm", "SNAP_DATE_date"])
    .groupby(["PN_norm", "SNAP_DATE_date"], as_index=False)["Qty Available"]
    .sum()
)

# merge inventory totals into rotabull rows on part + date
merged = Rotabull_df_copy_price.merge(
    inv_daily,
    left_on=["Part_norm", "Received_Date"],
    right_on=["PN_norm", "SNAP_DATE_date"],
    how="left",
    suffixes=("", "_inv")
)

# format Stock: numeric where found, 'NS' where not
def _format_stock(x):
    if pd.isna(x):
        return "NS"
    try:
        f = float(x)
    except Exception:
        return "NS"
    if f.is_integer():
        return int(f)
    return round(f, 2)

merged["Stock"] = merged["Qty Available"].apply(_format_stock)

# copy Stock back to Rotabull_df_copy_price and clean helper cols
Rotabull_df_copy_price["Stock"] = merged["Stock"].values

# optional: drop helper normalization columns if not needed
Rotabull_df_copy_price.drop(columns=["Part_norm"], inplace=True, errors="ignore")
DF_Inventory_copy.drop(columns=["PN_norm"], inplace=True, errors="ignore")

print(f"Added 'Stock' column to Rotabull_df_copy_price ({Rotabull_df_copy_price.shape[0]} rows). Sample:")
print(Rotabull_df_copy_price[["Part Number", "Received_Date", "Stock"]].head(10))
```

```python
# Single-Pass Fuzzy Matching (Pass 1 only)
# Removes Pass 2 to simplify matching while preserving downstream columns for compatibility.
import re
import pandas as pd
from rapidfuzz import process, fuzz

# ---------------------- Configuration ----------------------
MATCH_THRESHOLD = 80              # Threshold for accepting Pass 1 as "primary"

# ---------------------- Column Selection (Buyer Raw Name) ----------------------
BUYER_COL_CANDIDATES = [
    'Buyer Company Name', 'Buyer Company', 'Buyer Name', 'Company', 'Customer', 'Buyer_name_norm'
]
for _cand in BUYER_COL_CANDIDATES:
    if _cand in work_df.columns:
        buyer_name_col = _cand
        break
else:
    raise KeyError("No suitable buyer name column found for pass 1 matching.")

# Create a uniform raw text column for pass1 operations
work_df['buyer_name_pass1_raw'] = work_df[buyer_name_col].astype(str).fillna('')

# ---------------------- PASS 1: Direct Full-Name Matching ----------------------
# Minimal upper/strip normalization (NO generic filtering, NO suffix stripping)

def pass1_norm(s: str) -> str:
    if pd.isna(s):
        return ''
    return str(s).upper().strip()

customer_names_raw = customer_df['Customer'].astype(str).fillna('').tolist()
customer_names_raw_norm = [pass1_norm(x) for x in customer_names_raw]

buyer_unique_norm = pd.Series(work_df['buyer_name_pass1_raw'].apply(pass1_norm)).unique()

# Precompute list for search
choices_pass1 = customer_names_raw_norm

name_to_idx_p1 = {}
name_to_score_p1 = {}
for bname in buyer_unique_norm:
    if not bname:
        name_to_idx_p1[bname] = -1
        name_to_score_p1[bname] = 0
        continue
    res = process.extractOne(bname, choices_pass1, scorer=fuzz.WRatio, score_cutoff=0)
    if res is None:
        name_to_idx_p1[bname] = -1
        name_to_score_p1[bname] = 0
    else:
        # res -> (matched_string, score, index)
        name_to_idx_p1[bname] = int(res[2])
        name_to_score_p1[bname] = int(round(res[1]))

# Attach Pass 1 results
work_df['customer_match_pass1_idx'] = work_df['buyer_name_pass1_raw'].apply(lambda s: name_to_idx_p1.get(pass1_norm(s), -1)).astype('Int64')
work_df['customer_match_pass1_score'] = work_df['buyer_name_pass1_raw'].apply(lambda s: name_to_score_p1.get(pass1_norm(s), 0)).astype('Int64')
work_df['customer_match_pass1'] = work_df['customer_match_pass1_idx'].apply(lambda i: customer_names_raw[i] if i >= 0 and i < len(customer_names_raw) else pd.NA)

# ---------------------- Compatibility: stub Pass 2 outputs ----------------------
# Downstream cells expect these columns to exist; provide safe defaults
work_df['customer_match_pass2_idx'] = pd.Series([pd.NA] * len(work_df), dtype='Int64')
work_df['customer_match_pass2_score'] = pd.Series([pd.NA] * len(work_df), dtype='Int64')
work_df['customer_match_pass2'] = pd.Series([pd.NA] * len(work_df))
work_df['customer_match_pass2_detail'] = pd.Series(['pass2_removed'] * len(work_df))
work_df['customer_pass2_would_match'] = False

# ---------------------- Primary Merge (Using Pass 1 Only) ----------------------
# Prepare customer_df with prefix for safe merging
customer_prefixed = customer_df.add_prefix('customer_')

# Only accept pass1 matches that meet threshold
valid_primary_mask = (work_df['customer_match_pass1_score'] >= MATCH_THRESHOLD) & (work_df['customer_match_pass1_idx'] >= 0)
work_df['_cust_match_idx_primary'] = work_df['customer_match_pass1_idx'].where(valid_primary_mask, -1)

rotabull_enriched = work_df.merge(
    customer_prefixed,
    how='left',
    left_on='_cust_match_idx_primary',
    right_index=True,
)
rotabull_enriched.drop(columns=['_cust_match_idx_primary'], inplace=True)

# For backward compatibility: set 'customer_Customer' & 'customer_fuzzy_score' from Pass 1 accepted matches only
rotabull_enriched['customer_Customer'] = rotabull_enriched.get('customer_Customer').where(valid_primary_mask, pd.NA) if 'customer_Customer' in rotabull_enriched.columns else pd.NA
rotabull_enriched['customer_fuzzy_score'] = rotabull_enriched.get('customer_match_pass1_score').where(valid_primary_mask, pd.NA)

# Summary statistics
total_rows = len(rotabull_enriched)
pass1_hits = valid_primary_mask.sum()
print(f"Pass1 accepted (>= {MATCH_THRESHOLD}): {pass1_hits}/{total_rows}")
print("Columns added: customer_match_pass1, customer_match_pass1_score")

# Head preview
rotabull_enriched.head()
```
